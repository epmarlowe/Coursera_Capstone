### Then, I used a simple method to create my data frame

```python
d = pd.read_html('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M')
df = d[0]
df
```

### Then I renamed the columns

```python
df=df.rename(columns=df.iloc[0]).drop(df.index[0])
df
```

```python
list(df.columns)
```

### Then, I removed all rows where the Borough was "Not assigned"

```python
df = df[df.Borough != 'Not assigned']
df
```

### Then, I combined all rows where the Postcode was the same

```python
df = df.groupby(['Postcode' , 'Borough'])['Neighbourhood'].apply(' , '.join).reset_index()
df
```

```python
def comma_join(lit):
    return ','.join(lit)
```

```python
group = data=df.groupby(['Postcode', 'Borough'])['Neighbourhood'].apply(comma_join)
joinframe = pd.DataFrame(group, columns=['Neighbourhood'])
joinframe.reset_index(inplace=True)
```
